```python
counter = 0
for word in str(df_order_reviews['review_comment_message']).split():
    if 'excelente' in word.lower():
        counter += 1
```

```python
counter
```

```python
word_lst = {}
for word in str(df_order_reviews['review_comment_message']).split():
    word = word.lower()
    if word in word_lst:
        word_lst[word] += 1
    else:
        word_lst.update({word: 1})
```

```python
sorted(word_lst.items(), key=operator.itemgetter(1))
```
